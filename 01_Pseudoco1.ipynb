# Introducción a Psudocodigo

## ¿Qué es el pseudocódigo?

- Es una forma de representar algoritmos usando lenguaje natural estructurado.
- No sigue la sintaxis estricta de un lenguaje de programación.
- Permite planificar la lógica antes de programar.



- Usa estructuras similares a los lenguajes de programación.
- No es ejecutable por una computadora.
- Fácil de leer y entender.

### Elementos de un pseudocodigo
- Variables: Almacenan datos (números, texto, booleanos).
- Estructuras de control: Permiten decisiones y repeticiones.
- Entradas y salidas: Cómo el usuario interactúa con el programa.

### cruzar la calle
```
Algoritmo CruzarLaCalle

    es_seguro ← Falso
    Mientras no es_seguro hacer:
        Mirar a la izquierda y a la derecha
        Si no hay coches cerca entonces
            Cruzar la carretera
            es_seguro ← Verdadero
    cruzar
    Fin Mientras


Fin Algoritmo
```
### Preparar copmida
```
comida_lista ← Falso

Mientras no comida_lista hacer:
    Reunir ingredientes
    Si faltan ingredientes entonces
        Ir a la tienda y comprarlos
    Sino
        Preparar ingredientes (lavar, cortar, mezclar)
        Cocinar según la receta
        Si la comida está bien cocida entonces
            comida_lista ← Verdadero
            Servir la comida
        Sino
            Seguir cocinando y revisar de nuevo
Fin Mientras

Mostrar "¡La comida está lista para comer!"
```
